```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
world_cup = pd.read_csv('WorldCups.csv')
my_df = pd.DataFrame(world_cup)
my_df['goals_per_game'] = my_df['GoalsScored']/my_df['MatchesPlayed']
my_df
plt.plot(my_df['Year'],my_df['goals_per_game'],color='red')
plt.rc('lines', linewidth=1)
plt.ylabel('Goals')
plt.xlabel('Year')
plt.title('AGPG Over World Cup History')
### This graph takes the number of goals scored each year and divides it by the number of matches played to 
###return the average goals per match for each year in World Cup History 


```

```python
qb_stats = pd.read_csv('QBStats_all.csv')
plt.subplot(1, 2, 1)
td = pd.DataFrame(qb_stats)
pick = pd.DataFrame(qb_stats)
by_year_td =td.groupby('year')['td'].mean()
by_year_sack = pick.groupby('year')['sack'].mean()
plt.plot(by_year_td)
plt.ylabel('TouchDowns')
plt.xlabel('Year')
plt.title('QB TDPG 1995-2015')
plt.subplot(1, 2, 2)
plt.plot(by_year_sack)
plt.ylabel('Sacks')
plt.xlabel('Year')
plt.title('QB SAC PG 1995-2015')

### Comparison of how many TDS were thrown by each QB per game comapred to the average sacks per game. Could not inclued picks.
### Seems that the int - not coming through as an interger.
```
